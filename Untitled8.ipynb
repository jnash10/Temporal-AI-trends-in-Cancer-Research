```python
from serpapi import GoogleSearch
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pages = 10 #scrape the first 5 pages
keywords = ['Machine Learning', 'Neural Network', 'AI']
data = {'Machine Learning':[], 'Neural Network':[], 'AI':[]}
```

```python
for keyword in keywords:
  for i in range(pages):
    params = {
      "engine": "google_scholar",
      "q": f"{keyword} source:AACR",
      "api_key": "622bc4894297b0a8c4ab015c12fa38c7286c27e110171a5f9c3ec19e33398468",
      "start":f"{i+1}",
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    organic_results = results["organic_results"]
    for result in organic_results:
      data[keyword].append(result['publication_info']['summary'])
```

```python
d = pd.DataFrame(data)
pd.DataFrame.to_csv(d,'paper_info.csv')
d
```

```python
ML_years = {}
for i in range(2002, 2022):
  ML_years[i] = 0

for info in data['Machine Learning']:
  for year in ML_years.keys():
    if str(year) in info:
      ML_years[year]+=1
print(ML_years)

z = np.polyfit(list(ML_years.keys()), list(ML_years.values()), 1)
p = np.poly1d(z)

fig,ax = plt.subplots(1,1,figsize=(20,10))
ax.set_xticks(list(ML_years.keys()))

ax.plot(list(ML_years.keys()), list(ML_years.values()), label='values')
ax.plot(list(ML_years.keys()), p(list(ML_years.keys())), linestyle='dashed', label='trend')

plt.title('Research Paper with "Machine Learning" in title (2002-2021). ')
plt.xlabel('Year')
plt.ylabel('Research Papers')
plt.legend()
plt.savefig('ML.png')
plt.show()

d1 = pd.DataFrame(columns = ['Year', 'papers'])
d1['Year'] = list(ML_years.keys())
d1['papers'] = list(ML_years.values())
pd.DataFrame.to_csv(d1,'ML.csv')
d1
```

```python
fig,ax = plt.subplots(1,1,figsize=(20,10))
ax.plot(list(AI_years.keys()), list(AI_years.values()), label = 'AI')
ax.plot(list(NN_years.keys()), list(NN_years.values()), label = 'Neural Network')
ax.plot(list(ML_years.keys()), list(ML_years.values()), label = 'Machine Learning')

ax.set_xticks(list(ML_years.keys()))

plt.title('comparison')
plt.legend()
plt.savefig('combined.png')
plt.show()

d4 = pd.DataFrame(columns=['year','Machine Learning','Neural Network','AI'])
d4['year'] = list(AI_years.keys())
d4['Machine Learning'] = list(ML_years.values())
d4['Neural Network'] = list(NN_years.values())
d4['AI'] = list(AI_years.values())
pd.DataFrame.to_csv(d4,'combined.csv')
d4
```
